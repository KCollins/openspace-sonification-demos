This notebook's purpose is to generate .lua files for ground-based stations, based on the example file from  https://docs.openspaceproject.com/en/latest/generated/asset-components/GlobeTranslation.html:

## Import station lat/lon data

```python
import pandas as pd
stations = pd.read_csv("stations.csv")
stations
```

## Import template
In the future we might want to do something more clever to parse this Lua file, but for now I'm mainly just pulling in the file as text and swapping the relevant strings with a brute force approach in the next section below.

```python
from luaparser import ast
# from luaparser import astnodes

with open('template.lua', 'r') as file:
    src = file.read()

tree = ast.parse(src)
# print(ast.to_pretty_str(tree))
foo = tree.body.to_json()

# class NumberVisitor(ast.ASTVisitor):
#     def visit_Number(self, node):
#         print('Number value = ' + str(node.n))

# tree = ast.parse(src)
# NumberVisitor().visit(tree)
```

## Generate and save .lua files for stations

```python
for index, row in stations.iterrows():
    IAGA = row['IAGA']
    LAT = row['GEOLAT']
    LON = row['GEOLON']
    print(row['IAGA'], row['GEOLAT'], row['GEOLON'])
    foo = src.replace("Latitude = 20.0", "Latitude = " + str(LAT))
    bar = foo.replace("Longitude = -45.0", "Longitude = " + str(LON))
    print(bar,  file=open('Position/' + IAGA + '.lua', 'w'))
```
